## Exercise 2 - Step 1: Read CSV using pandas. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

healthcare = pd.read_csv('healthcare_patients.csv')
print("\nHealthcare Data:")
print(healthcare)
```

## Exercise 2 - Step 2: Define IF-THEN rules for symptoms to diagnosis mapping. 

```python
# Simple rule-based diagnosis
def diagnose(symptoms):
    if "Fever" in symptoms:
        return "Possible Infection"
    elif "Chest Pain" in symptoms:
        return "Seek Cardiac Evaluation"
    elif "Headache" in symptoms:
        return "Possible Migraine"
    elif "Cough" in symptoms:
        return "Respiratory Check Recommended"
    else:
        return "Monitor Symptoms"
```

## Exercise 2 â€“ Step 3: Apply rules and create a new column 'SuggestedDiagnosis'.

```python
healthcare['SuggestedDiagnosis'] = healthcare['Symptoms'].apply(diagnose)
print("\nDiagnosis Suggestions:")
print(healthcare)
```

```python
# Optional visualization: count diagnoses
plt.figure(figsize=(8,5))
sns.countplot(y="SuggestedDiagnosis", data=healthcare, palette="crest")
plt.title("Suggested Diagnoses Frequency")
plt.xlabel("Count")
plt.ylabel("Diagnosis Type")
plt.show()
```
